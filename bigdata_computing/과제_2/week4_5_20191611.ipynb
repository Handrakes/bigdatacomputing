```python
import csv
import math
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'AppleGothic'##한글 설정 in Mac
plt.rcParams['axes.unicode_minus'] = False

f = open('/Users/handrakes/bigdata/gender.csv', 'r', encoding="cp949")
data = csv.reader(f)
m = []
f = []
arr = input('찾고 싶은 지역의 이름을 알려주세요 : ')
for row in data:
    if arr in row[0]:
        for i in range(0, 101):
            m.append(int(row[i+3].replace(',','')))
            f.append(int(row[-(i+1)].replace(',','')))
        break    
f.reverse()

size = []
for i in range(0,101):
    size.append(float(math.sqrt(m[i]+f[i])))
#print(size)

plt.figure(figsize = (10,5))
plt.style.use("ggplot")
plt.title(arr + "지역의 성별 인구 그래프")
plt.scatter(m,f,s = size,c=range(101), alpha = 0.7, cmap='jet')
plt.plot(range(max(m)),range(max(m)), 'g')
plt.xlabel("남성 인구 수")
plt.ylabel("여성 인구 수")
plt.colorbar()
plt.show()
```

```python
import numpy as np
import csv
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'AppleGothic'##한글 설정 in Mac
plt.rcParams['axes.unicode_minus'] = False ##

f = open('/Users/handrakes/bigdata/age_no_comma.csv', 'r', encoding = 'cp949')
data = csv.reader(f)
next(data)
data = list(data)

mn = 1 # 최대값을 저장할 변수 생성 및 초기화
result_name = '' # 최대값을 갖는 지역의 이름을 저장할 변수 생성 및 초기화
result = 0 # 최대값을 갖는 지역의 연령대별 인구 비율을 저장할


for row in data :
    if int(row[2]) == 0:
        continue
    tmp = np.array(row[3:10], dtype = 'int64')
    temp = np.array(row[3: ], dtype = 'int64')
    s = tmp.sum() / int(row[2]) * 100 
    if s > mn:
        mn = s
        result_name = row[0]
        result = temp

print(result_name + "지역의 0세 이상 6세 이하의 인구 비율: %.1f" %(mn) + "%")
plt.figure(figsize = (10,5))
plt.style.use("ggplot")
plt.title(result_name + " 지역의 인구 구조")
plt.plot(result)
plt.show()
```
