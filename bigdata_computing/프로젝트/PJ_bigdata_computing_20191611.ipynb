```python
#커리의 기록
import pandas as pd
import matplotlib.pyplot as plt

url = 'https://www.espn.com/nba/player/stats/_/id/3975/stephen-curry'
tables = pd.read_html(url)

df = pd.concat([tables[0], tables[1]], axis = 1).dropna() #2개의 데이터를 하나로 묶어서 보기
df
```

```python
#여기서 부상이었던 2019 시즌은 제외 
df.drop([10, ], axis = 0, inplace = True)
df
```

```python
#index 초기화하기
df.reset_index(inplace=True)
df.drop(['index'], axis = 1, inplace = True)

plt.rcParams['font.family'] = 'AppleGothic' ##한글 설정 in Mac
plt.rcParams['axes.unicode_minus'] = False

season = df['season']
points = df['PTS'] #평균득점
fg = df['FG%'] #야투율

#커리의 나이를 생각했을 경우 점점 올라가는 그래프
plt.figure(figsize = (18,9))
plt.plot(season, points, lw = 5)
plt.xlabel('Season', size = 20)
plt.ylabel('Season', size = 20)
plt.title('평균 득점 추이', size = 40)
plt.grid()
plt.show()
```

```python
#커리의 야투중 3점 야투의 비율 구하기
df['FGs'] = df['FG'].str.split('-').str[0].astype(float) #야투 시도 갯수
df['3PT FGs'] = df['3PT'].str.split('-').str[0].astype(float) #3점 야투 시도 갯수
df['3PT portion'] = df['3PT FGs'] / df['FGs'] # 3점 야투가 차지하는 비율

plt.figure(figsize = (18,10))
plt.plot(season, df['3PT portion'])
plt.show()
#커리가 전보다 점점 더 많은 3점을 쏘고 있다!!를 알 수 있음
```

```python
#커리의 180클럽 기록을 계산해보자
#180클럽은 농구에서 슈터로써의 능력을 평가하는 지표이다. 
#2점 야투율 50%이상, 3점 야투율 40%이상, 자유투 90%이상
#현재까지 180클럽을 성공한 선수는 총 11명인데, 그 중 평균 득점 30점을 넘긴 선수는 커리가 유일

#2점 야투율 계산하기
pt2_success = df['FG'].str.split('-').str[0].astype(float) - df['3PT'].str.split('-').str[0].astype(float)
pt2_all = df['FG'].str.split('-').str[1].astype(float) - df['3PT'].str.split('-').str[1].astype(float)
df['2P%'] = (pt2_success / pt2_all) * 100


df['180club'] = df['3P%'] + df['FT%'] + df['2P%']
plt.figure(figsize = (18,10))
plt.plot(season, df['180club'], lw = 10)
plt.title("180클럽 달성 추이", size = 40)
plt.grid()
#plt.show()
#180 클럽에 가입하려면 최소 야투성공 300개, 3점 성공 82개, 자유투 성공 125개 물론 이 조건을 충족하는 시즌이 15-16시즌 밖에 없기는 하다. 
```

```python
from nba_api.stats.static import players
import pandas as pd


#NBA 선수들 리스트 가져오기
nba = players.get_players()
name = []
ids = []

#현역 선수만 포함
for i in nba:
    if i['is_active'] == True:
        name.append(i['full_name'])
        ids.append(i['id'])
        
dff = pd.DataFrame([name, ids]).T
dff.columns = ['name', 'id']
dff['name'] = dff['name'].str.lower()

sga = dff.query('name == "shai gilgeous-alexander"')['id'].values[0]
```

```python
#TS%(True shooting) : 얼마나 효율적으로 공격을 하였는가
#TS = (총 득점) / [2 *{야투시도 + (0.44 * 자유투 시도)}]
#TS = PTS /  

career = playercareerstats.PlayerCareerStats(player_id = curry)
temp = career.get_data_frames()
temp[0]
```

```python
temp
temp[0]['TS%'] = (temp[0]['PTS']) / (2 * (temp[0]['FGA'] + (0.44 * temp[0]['FTA'])))
temp[0]
```

```python
#sga의 TS 계산
career_sga = playercareerstats.PlayerCareerStats(player_id = sga)
temps = career_sga.get_data_frames()
temps[0]
```

```python
# nba 평균 선수들 기록을 구하기 위해서 basketball-reference에서 정보 추출

import requests
from bs4 import BeautifulSoup
from tabulate import tabulate

# Send a GET request to the webpage URL
url = "https://www.basketball-reference.com/leagues/NBA_2023_totals.html"
response = requests.get(url)

# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')

# Find the table element containing the data
table = soup.find('table', {'id': 'totals_stats'})

# Find all rows in the table
rows = table.find_all('tr')

# Extract the headers from the first row
headers = [header.get_text() for header in rows[0].find_all('th')]

# Create an empty list to store player records
player_records = []

# Iterate over the rows and extract the player records
for row in rows[1:]:
    # Extract the data from each column
    data = [col.get_text() for col in row.find_all('td')]
    
    # Append the data to the player records list
    player_records.append(data)

    
all = pd.DataFrame(player_records)

all
```
